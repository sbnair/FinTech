```python
from bokeh.io import output_notebook, show
from bokeh.layouts import gridplot
from bokeh.plotting import figure

output_notebook()


x = final["GEI Score"]
y0 = final["ATMs per 1000 Adults"]
y1 = final['iPhone Users']
y2 = final["Starbucks Locations"]

s1 = figure(width=300, plot_height=300, title="ATMs per 1000 Adults")
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

s2 = figure(width=300, height=300, title='iPhone Users')
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)


s3 = figure(width=300, height=300, title='Starbucks Locations')
s3.square(x, y2, size=10, color="orange", alpha=0.5)

p = gridplot([[s1, s2, s3]],toolbar_location=None)

show(p)
```

```python
df_1 = final[["Country", "GEI Score"]].sort_values("GEI Score", ascending=1)
from bokeh.charts.attributes import CatAttr


defaults.plot_width = 900
defaults.plot_height = 400

bar_plot_2 = Bar(df_1,values='GEI Score', label=CatAttr(columns=['Country'],sort=False),title="GEI Score", color="green", legend=None, toolbar_location=None)
bar_plot_2.title.text_font_size = '10pt'
#plt = Bar(data, values='values', label=CatAttr(columns=['labels'], sort=False),)
output_notebook()
show(bar_plot_2)
```
